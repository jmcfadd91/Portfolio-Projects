```python
df =  pd.read_csv('/Users/James/Desktop/LIQUOR STORE STUFF/TransactionDec21.csv')
df.head(5)
```

```python
df['Unit Retail'] = df['Unit Retail'].str.replace('$','')
df['Unit Retail'] = df['Unit Retail'].str.replace(',','')
df['Unit Retail'] = df['Unit Retail'].astype(float)
```

```python
items = df.groupby(['Item Desc & Size'])['Unit Retail'].max().to_dict()
df1 = pd.DataFrame(items, index=[0]).T.reset_index().rename(columns={'index': 'product_name', 0: 'product_price'})
df1['product_id'] = df1.index + 1
df1.map(lambda x: x)
```

```python
df1.merge(df, left_on='product_name', right_on='Item Desc & Size', how='right').drop(columns=['Trans ID', 'Item Descr', 'Retail', 'Units', 'Unit Retail', 'Item Desc & Size'])
```
